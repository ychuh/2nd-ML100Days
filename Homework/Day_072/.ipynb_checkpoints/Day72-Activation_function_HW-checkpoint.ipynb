- 選用啟動函數, 主要是參考幾個:
  
  (1) 避免梯度消失gradient vanishing 

  (2) 避免梯度爆炸

  (3) 函數輸出可以是zero-centered 

  (4) 進行指數運篹

```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
# Relu Function
def relu(x):
    return list(map(lambda x:x if x>0 else 0,x))

# Derived dReLu function
def drelu(x):
    return list(map(lambda x: 1 if x>0 else 0,x))
```

```python
x = np.arange(-10,10,0.01)
y = relu(x)

#ReLU函数图像
plt.plot(x,y)
plt.grid(color='b' , linewidth='0.1' ,linestyle='--')
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(1))
plt.show()
```

```python
#ReLU函数导数图像
dev_y = drelu(x)
plt.plot(x,dev_y)
plt.grid(color='b' , linewidth='0.1' ,linestyle='--')
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(0.1))
plt.show()
```
