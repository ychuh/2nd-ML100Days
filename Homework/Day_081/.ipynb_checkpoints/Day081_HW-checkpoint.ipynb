```python
import os
import keras

# Disable GPU
os.environ["CUDA_VISIBLE_DEVICES"] = ""
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()
```

```python
## 資料前處理
def preproc_x(x, flatten=True):
    x = x / 255.
    if flatten:
        x = x.reshape((len(x), -1))
    return x

def preproc_y(y, num_classes=10):
    if y.shape[-1] == 1:
        y = keras.utils.to_categorical(y, num_classes)
    return y    
```

```python
# Preproc the inputs
x_train = preproc_x(x_train)
x_test = preproc_x(x_test)

# Preprc the outputs
y_train = preproc_y(y_train)
y_test = preproc_y(y_test)

x_train.shape
```

```python
from keras.regularizers import l1, l1_l2

def build_mlp(input_shape, regularizer, ratio, outputs=10, num_nurons=[512, 256, 256, 128, 128]):
    
    input_layer = keras.layers.Input(input_shape)
    
    for i, units in enumerate(num_nurons):
        if i==0:
            x = keras.layers.Dense(units=units, activation="relu", kernel_regularizer=regularizer(ratio), name="Input_layer")(input_layer)
        else:
            x = keras.layers.Dense(units=units, activation="relu", kernel_regularizer=regularizer(ratio), name="hidden_layer%s" %str(i))(x)
    
    output_layer = keras.layers.Dense(units=10, activation="softmax", name="output_layer")(x)
    model = keras.models.Model(inputs=[input_layer], outputs=[output_layer])
    
    return model
```

```python
batch_size = 256
epochs = 50
lr = 1e-3
regularizer = [l1, l1_l2]
reg_keys = ['l1', 'l1_l2']
ratios = [1e-2, 1e-4, 1e-8, 1e-12]
sgd = keras.optimizers.SGD(lr=lr, momentum=0.9, nesterov=True)
```

```python
results = {}

for i, reg in enumerate(regularizer):
    for r in ratios:
        
        model = build_mlp(x_train.shape[1:], reg, r)
        model.compile(optimizer=sgd, loss='categorical_crossentropy', metrics=['accuracy'])
        history = model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs, verbose=1, shuffle=True, validation_data=(x_test, y_test))
        
        # Collecting accuracy/loss
        train_loss = history.history['loss']
        test_loss = history.history['val_loss']
        train_acc = history.history['acc']
        test_acc = history.history['val_acc']
        
        tag = "%s -- lr: %s" %(reg_keys[i], str(r))
        results[tag] = {"train_loss": train_loss,
                        "test_loss": test_loss,
                        "train_acc": train_acc,
                        "test_acc": test_acc
                       }
        
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
cmap = plt.get_cmap('gnuplot')
colors = [cmap(i) for i in np.linspace(0, 1, len(results))]

plt.figure(figsize=(10, 6))
for i, cond in enumerate(results.keys()):
    plt.plot(results[cond]['train_acc'], '-', label=cond, color=colors[i])
    plt.plot(results[cond]['test_acc'], '--', label=cond, color=colors[i])
plt.title("Accuracy")
plt.legend(loc="upper left")
plt.show()

plt.figure(figsize=(10, 6))
for i, cond in enumerate(results.keys()):
    plt.plot(range(len(results[cond]['train_loss'])),results[cond]['train_loss'], '-', label=cond, color=colors[i])
    plt.plot(range(len(results[cond]['test_loss'])),results[cond]['test_loss'], '--', label=cond, color=colors[i])
plt.title("Loss")
plt.legend(loc="upper right")
plt.show()
```

```python
loss, acc = model.evaluate(x_test, y_test)
```

```python
print(f'Accuracy: {acc}')
print(f'Loss: {loss}')
```
